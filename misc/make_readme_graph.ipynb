```python
%load_ext hierarchymagic
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from hierarchymagic import run_dot
from IPython.display import Image
```

```python
cmd = """\
diGraph G{labelloc="t"; label="";
          subgraph cluster  {
          style=invis
          graph [shape=circle];
          node [color=lightblue2, style=filled]; 
          
          Notebook_Master[color=lightgray];
        
          Notebook_Mother[color=lightgray];
          Slides_Notebook[color=lightgray];
          PDF_Notebook[color=lightgray];
          HTML_Notebook[color=lightgray];
        
          Slides_file[color=lightgray];
          PDF_file[color=lightgray];
          HTML_file[color=lightgray];
        
          Notebook_Master->Notebook_Mother[label=" cherry_pick: \\n 'not rough_notes'", fontsize=10];
          
          Notebook_Mother->PDF_Notebook[label=" cherry_pick: \\n 'pdf_only and (not omit_pdf)'", fontsize=10];
          Notebook_Mother->Slides_Notebook[label=" cherry_pick: \\n 'slides_only and (not omit_slides)'", fontsize=10];
          Notebook_Mother->HTML_Notebook[label=" cherry_pick: \\n 'html_only and (not omit_html)'",fontsize=10];
          
          PDF_Notebook->PDF_file[label=" nbconvert: \\n ' --to latex --post PDF --template ...'", fontsize=10]
          Slides_Notebook->Slides_file[label=" nbconvert: \\n '--to slides --template ...'", fontsize=10]
          HTML_Notebook->HTML_file[label=" nbconvert: \\n ' --to html '", fontsize=10];
          
           }
          }
"""
```

```python
tmpfile = 'readme_graph.png'
with open(tmpfile, 'w') as f: f.write(run_dot(cmd))
```

```python
Image(tmpfile,width="100%")
```
