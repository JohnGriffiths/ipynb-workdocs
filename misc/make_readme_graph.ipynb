```python
%load_ext hierarchymagic
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from hierarchymagic import run_dot
from IPython.display import Image
```

```python
cmd = """\
digraph G {

subgraph cluster_all{
		color=white;
		label = "The ipynb-workdocs production line\\n \\n"
		fontsize=30
		fontname="times bold italic"        
		fontcolor="black"     
        

subgraph cluster_rolling{
		label = "- Rolling doc \\n  - Personal copy";
		labeljust="r"
		fontname="times italic"
		fontcolor="black"     
		fontsize=15
		color=black
		Mas_nb[style=filled,color=lightgrey,label="Master Notebook \\n(.ipynb)"];
	}
     
    
subgraph cluster_timestamp{
		node [style=filled];
		label = " - Latest version (DD/MM/YYYY)            \\n- Shared copy (rough notes removed)     \\n- Multiple formats for different use-cases"
		fontname="times italic"        
		labeljust="r"
		fontcolor="black"        
		fontsize=15        
		color=black

Mum_nb[style=filled, color=lightgrey,label="Mother Notebook \\n(.ipynb)"];

Slides_nb[color=lightgray, label = "Slides Notebook \\n(.ipynb)"];
PDF_nb[color=lightgray, label = "PDF Notebook \\n(.ipynb)"];
HTML_nb[color=lightgray, label = "HTML Notebook \\n(.ipynb)"];
        
Slides_html_file[color=lightgray,label="Reveal slides \\n(.slides.html)"];
Slides_pdf_file[color=lightgray, label="PDF of Reveal slides \\n(.pdf)"];
PDF_file[color=lightgray, label = "PDF file \\n(.pdf)"];
HTML_file[color=lightgray, label = "HTML file \\n(.html)"];

Mas_nb -> Mum_nb[label=" cherry_pick: \\n 'not rough_notes'", fontsize=10,color=grey];

Mum_nb -> PDF_nb[label=" cherry_pick: \\n 'pdf_only and (not omit_pdf)'", fontsize=10, color=grey];
Mum_nb -> Slides_nb[label=" cherry_pick: \\n 'slides_only and (not omit_slides)'", fontsize=10, color=grey];
Mum_nb -> HTML_nb[label=" cherry_pick: \\n 'html_only and (not omit_html)'",fontsize=10, color=grey];



	subgraph cluster_slides {
		node [style=filled];
		Slides_nb -> Slides_html_file[label=" nbconvert: \\n '--to slides --template ...'", fontsize=8]
		Slides_html_file -> Slides_pdf_file[label="Save as PDF \\n from browser", fontsize=10]
		label = "Slides";
		labeljust = "c"        
		fontname="times italic"
		fontsize=15
		fontcolor="orange"        
		color=orange;
	}

	subgraph cluster_pdf {
		PDF_nb -> PDF_file[label=" nbconvert: \\n ' --to latex --post PDF --template ...'", fontsize=9]
		label = "PDF Document";
		labeljust = "c"                
		fontname="times italic"
		fontsize=15
		fontcolor="red"        
		color=red
	}
    
	subgraph cluster_html {
		color=lightblue;
		HTML_nb -> HTML_file[label=" nbconvert: \\n ' --to html '", fontsize=9];
		label = "HTML Document";
		labeljust = "c"        
		fontname="times italic"
		fontsize=15
		fontcolor="blue"                
		color=blue
	}
    

}          

}

}"""
```

```python
tmpfile = 'readme_graph.png'
with open(tmpfile, 'w') as f: f.write(run_dot(cmd))
```

```python
Image(tmpfile,width="100%")
```
