# Dealing with multiple notebooks

A common situation is to have documentation spread across multiple notebooks. This may be because the content is naturally broken up into independent sections, or simply to make files more manageable and/or less likely to crash the browser, which extremely large notebooks are prone to doing. 



We can still use the ipynb-workdocs workflow with multiple notebooks, with a few minor adjustments. 

First it is best to have a separate 'organizer' notebook that does the function calls. These calls will include concatenating the separate notebooks into one single file, which is used to generate the PDF. 




The main point to bear in mind is that the table of contents (TOCs) that are marked to keep in the HTML documents need to reflect the intendedn usage of the files. If the intention is to have separate webpages and/or nbviewer pages for each of the notebooks, then no modifications are needed. If however you want a single HTML/nbviewer page for all of the notebooks, and a corresponding TOC, then you will need to make some manual adjustments to the concatenated file (see below) after nbconvert. This should only take a minute or so. The individual HTML/nbviewer files with their own TOCs are also generated and kept along the way, so it is straightforward to do both. Note that the TOC for the PDF will be generated from the concatenated notebook. 

    



1. Nbconvert to html each notebook
2. Concatenate
3. Nbconvert to html the concatenated one
3. Nbconvert to slides the concatenated one
4. Nbconvert to pdf the concatenated one
